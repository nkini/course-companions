Applicable when the covariance both classes are Gaussian distributed, with same covariance but different means.

$ w^T = (\mu_1^T - \mu_0^T) \Sigma^{-1} $

$ b = \frac{1}{2} \mu_0^T \Sigma^{-1}  \mu_0 
      - \frac{1}{2} \mu_1^T \Sigma^{-1} \mu_1 
      - ln \frac{\alpha}{1 - \alpha}$
      
- Also called Linear Discriminant Analysis
- If covariance is a diagonal matrix: Naive Bayes
- If we have different covariances: Quadratic Discriminant Analysis
- Optimal solution provided assumptions about the distributions are true.
