If we have a polynomial, we can still use linear regression

The linear means the weight parameters are linear

It says nothing about what $X$ has to be

Think about $aX^2 + bX + c$...  
This is linear in $a, b$ and $c$... with $X^2, X, 1$ as the coefficients, get it?

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d, Axes3D
%matplotlib inline

X = []
Y = []

for line in open('../machine_learning_examples/linear_regression_class/data_poly.csv'):
    x, y = line.split(',')
    x, y = float(x), float(y)
    X.append([1, x, x**2])
    Y.append(y)
```

```python
X, Y = np.array(X), np.array(Y)

fig = plt.figure()
ax = Axes3D(fig)

ax.scatter(X[:,1], X[:,2], Y)
plt.show()
```

```python
plt.scatter(X[:,1], Y)
plt.show()
```

```python
plt.scatter(X[:,2], Y)
plt.show()
```

```python
w = np.linalg.solve(np.dot(X.T,X), np.dot(X.T,Y))
Y_hat = np.dot(X, w)

plt.scatter(X[:,1], Y)
plt.plot(sorted(X[:,1]), sorted(Y_hat))
```

```python
d1 = Y - Y_hat
d2 = Y - Y.mean()
R2 = 1 - d1.dot(d1)/d2.dot(d2)
R2
```
