```python
# Sampling from a Gaussian PDF
```

```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Find prob density of 0 from the normal distribution
norm.pdf(0)
```

```python
# mean and variance other than 0, 1
# scale is std, loc is mean
norm.pdf(0, loc=5, scale=10)
# Variance is 100 if std is 10
```

```python
r = np.random.randn(10)
norm.pdf(r)
```

```python
# log of the pdf 
# Because adding is cheaper than multiplication. Also, no exponentiation.
# Working with log probabilities is much better than working with the pdf.
norm.logpdf(r)
```

```python
norm.cdf(r)
```

```python
norm.logcdf(r)
```

```python
r = np.random.randn(10000)
plt.hist(r, bins=100)
plt.show()
```

```python
r = 10 * np.random.randn(10000) + 5
# std = 10, mean = 5
plt.hist(r, bins=100)
plt.show()
```

```python
# Multidimensions - 2D Gaussian 
#(spherical Gaussian)
r = np.random.randn(10000, 2)
plt.scatter(r[:,0], r[:,1])
plt.show()
```

```python
#(elliptical Gaussian)
r[:, 1] = 5 * r[:, 1] + 2
plt.scatter(r[:,0], r[:,1])
plt.show()
# Notice that the axes are adjusted to balance the data
```

```python
plt.axis('equal')
plt.scatter(r[:,0], r[:,1])
plt.show()
```

```python
from scipy.stats import multivariate_normal as mvn
# Sampling from a general multivariate distribution, where dimensions are not necessarily independent from one another. (A full covariance matrix)
cov = np.array([[1, 0.8], [0.8, 3]])
# (A variance of 1 in the first dimension, a variance 3 in the second dimension, and a covariance of 0.8 between the dimensions)
mu = np.array([0,2])
r = mvn.rvs(mean=mu, cov=cov, size=1000)
```

```python
plt.scatter(r[:, 0], r[:, 1])
plt.axis('equal')
plt.show()
```

```python
r = np.random.multivariate_normal(mean=mu, cov=cov, size=1000)
plt.scatter(r[:, 0], r[:, 1])
plt.axis('equal')
plt.show()
```

Oddly, FFT is in Numpy and not in Scipy

```python
x = np.linspace(0, 100, 10000)
y = np.sin(x) + np.sin(3*x) + np.sin(5*x)
plt.plot(y)
plt.show()
```

```python
Y = np.fft.fft(y)
plt.plot(np.abs(Y))
plt.show()
```
