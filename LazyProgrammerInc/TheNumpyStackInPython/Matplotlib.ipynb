```python
# One assumption we make when doing linear regression is that the error is normally distributed. So if we take all the y data points and subtract it from its corresponding place in the predicted line, that data should be normally distributed.
y_actual = 2*x + 1
residuals = y - y_actual
plt.hist(residuals)
plt.show()
```
